```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# 1. Dataset Creation
data = {
    "Movie_Name": [
        "Inception",
        "The Dark Knight",
        "Interstellar",
        "Avatar",
        "Avengers: Endgame",
        "Spiderman",
        "Leo",
        "Jailer",
    ],
    "Genre": [
        "Sci-Fi",
        "Action",
        "Sci-Fi",
        "Sci-Fi",
        "Action",
        "Action",
        "Action",
        "Action",
    ],
    "Rating": [8.8, 9.0, 8.6, 7.8, 8.4, 7.3, 7.5, 7.2],
    "Votes": [2000, 2500, 1800, 1500, 2200, 1200, 950, 850],
}
df = pd.DataFrame(data)

print("=================== PROJECT 1: MOVIE DATASET ===================")
print(df)
print("\n" + "=" * 60)

# 2. Statistical Analysis
print(f"[*] Average Movie Rating : {df['Rating'].mean():.2f} / 10")
highest_rated = df.loc[df["Rating"].idxmax()]
print(
    f"[*] Highest Rated Movie  : {highest_rated['Movie_Name']} ({highest_rated['Rating']})"
)
print("=" * 60 + "\n")

# 3. Data Visualization
plt.figure(figsize=(10, 5))
sns.barplot(x="Movie_Name", y="Rating", data=df, palette="viridis")
plt.title("Movie Ratings Comparison Analysis", fontsize=12, fontweight="bold")
plt.xlabel("Movie Titles")
plt.ylabel("IMDb Ratings")
plt.xticks(rotation=30)
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```
